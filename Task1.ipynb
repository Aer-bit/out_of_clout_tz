```python
import numpy as np
import pandas as pd
```

```python
reviews = pd.read_csv('reviews.csv', sep=';', encoding='windows-1251')
```

```python
reviews.head()
```

```python
features = ['id клиента', 'дата создания', 'оценка заказа']
```

```python
reviews_clean = reviews[features]
```

```python
reviews_clean.sample(20)
```

```python
reviews_clean.shape
```

```python
reviews_clean.info()
```

```python
reviews_clean.dropna(inplace=True)
```

```python
reviews_clean.info()
```

```python
reviews_clean.sample(30)
```

```python
reviews_clean['дата создания'] = pd.to_datetime(reviews_clean['дата создания'], errors='coerce')
```

```python
def map_reviews(review):
    if review == '1':
        return 1
    elif review == '2':
        return 2
    elif review == '3':
        return 3
    elif review == '4':
        return 4
    elif review == '5':
        return 5
    else:
        return None
```

```python
reviews_clean['оценка заказа'] = reviews_clean['оценка заказа'].apply(lambda x: map_reviews(x))
```

```python
reviews_clean['оценка заказа'] = reviews_clean['оценка заказа'].apply(lambda x: int(x))
```

```python
reviews_clean['id клиента'] = reviews_clean['id клиента'].apply(lambda x : int(x))
```

```python
reviews_clean.dropna(inplace=True)
```
